```python
class tag:
    def __init__(self,_name,*body,**attr):
        self.name = _name
        self.body = body
        self.attr = {a:attr[a] for a in attr if attr[a]!=None
                    }
    
    def __repr__(self):
        attr = ''.join(map(lambda x:' {}={} '.format(*x),self.attr.items()))
        body = '\n'.join(map(str,self.body))
        if body:
            body+=''
        if self.name==None:
            return '''{body}'''.format(body=body)
        body = body.replace('\n','\n  ')

        ans = '''<{name}{attr}>\n  {body}\n</{name}>'''.format(name=self.name,body=body,attr=attr)
        return ans
    def __gt__(self,o):
        return tag(self.name,*self.body+(o,),**self.attr)
    def __add__(self,other):
        return tag(None,self,other)
    def __mul__(self,n):
        return tag(None,*((self,)*n) )

def h1(*body,**attr):
    return tag('h1',*body,**attr)

def h2(*body,**attr):
    return tag('h1',*body,**attr)

def h3(*body,**attr):
    return tag('h1',*body,**attr)

def h4(*body,**attr):
    return tag('h1',*body,**attr)

def a(*body,**attr):
    return tag('a',*body,**attr)

def tr(*body,**attr):
    return tag('tr',*body,**attr)

def th(*body,**attr):
    return tag('th',*body,**attr)
def label(*body,**attr):
    return tag('lable',*body,**attr)

def button(*body,**attr):
    return tag('button',*body,**attr)


def form(*inputs,action='/',method='GET'):
    from functools import reduce
    if inputs:
        return tag(
    'form',action='/tree',method='POST')>reduce(lambda x,y:x+y,inputs)
    return tag(
    'form',action='/',method=method)

def textInput(name='',default='',**kwargs):
    return tag('input',name=name,**kwargs)

def table(matrix):
    from functools import reduce
    ans = tag('table')
    lines = reduce(lambda x,y:(x)+(y),map(tr,[reduce(lambda x,y:x+y,[th(i) for i in line ]) for line in matrix]),)
    return ans>lines
fields =['surname','name','age']
Lorem_imsum='''Lorem Ipsum is simply dummy text of the printing and typesetting industry. Lorem Ipsum has been the industry's standard dummy text ever since the 1500s, when an unknown printer took a galley of type and scrambled it to make a type specimen book. It has survived not only five centuries, but also the leap into electronic typesetting, remaining essentially unchanged. It was popularised in the 1960s with the release of Letraset sheets containing Lorem Ipsum passages, and more recently with desktop publishing software like Aldus PageMaker including versions of Lorem Ipsum.
'''
HTML(repr(
    h1('Example Table:')+
    table([[2,4,'jsdfds'],[4,3,5]])+
    h3('Basic Form:')+
    form(
       *[ label('{}: '.format(f),For=f)+textInput(name=f,placeholder=f)
         for f in fields]
        +[button('Submit')],
        )
    >tag('p',Lorem_imsum)
    )
    )
```
